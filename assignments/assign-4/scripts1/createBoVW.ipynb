```python
import numpy as np
import matplotlib.pyplot as plt
import os, argparse, math, random
from datetime import datetime
import errno

start_time = datetime.now()

outputPath = "../dataset/prep/2b/bovw/"

if not os.path.exists(os.path.dirname(outputPath)):
    try:
        os.makedirs(os.path.dirname(outputPath))
    except OSError as exc:  # Guard against race condition
        if exc.errno != errno.EEXIST:
            raise

inputPath = "../../assign-2/prepData/dataset/group08_1/"

meanPath = "../dataset/prep/2b/bovw/imageDATA.kmeans"

classNames = ["bayou", "chalet", "creek"]
dataTypes = ["test", "train"]

# testPath = []
# testPath.append("../../assign-2/prepData/dataset/group08_1/test/bayou")
# testPath.append("../../assign-2/prepData/dataset/group08_1/test/chalet")
# testPath.append("../../assign-2/prepData/dataset/group08_1/test/creek")

# trainPath = []
# trainPath.append("../../assign-2/prepData/dataset/group08_1/train/bayou")
# trainPath.append("../../assign-2/prepData/dataset/group08_1/train/chalet")
# trainPath.append("../../assign-2/prepData/dataset/group08_1/train/creek")

data = []
means = []
```

```python
def fileHandle(fileName):
    wholeData = []
    file = open(fileName)
    for line in file:
        teLine = line.rstrip('\n ').split(' ')
        nLine = [float(i) for i in teLine]
        nLine = np.array(nLine)
        wholeData.append(nLine)
    file.close()
    return wholeData

means = fileHandle(meanPath)

def createBovW(dataType, className):
    bowVect = []
    
    for file in data:
        bow = np.zeros(len(means))
        for point in file:
            distances = np.zeros(len(means))
            for i in range(len(means)):
                distances[i] = np.linalg.norm(point-means[i])
            bow[int(np.argmin(distances))] += 1
        bowVect.append(bow)
        
    print("Data Length: ", len(data), " | BoVW length: ", len(bowVect))
     
    path = outputPath + dataType + "_" + className + ".bovw"
    
    if not os.path.exists(os.path.dirname(path)):
        try:
            os.makedirs(os.path.dirname(path))
        except OSError as exc:  # Guard against race condition
            if exc.errno != errno.EEXIST:
                raise
    try:
        print("target File: ", path)
        outfile = open(path, "w")
    except IOError:
        print("File not created !!!!!!!!!!!!!!!!!!!!!!!!!")
        
    for line in bowVect:
        for feature in line:
            outfile.write(str(feature)+" ")
        outfile.write("\n")
        
    outfile.close()
```

```python
for dataType in dataTypes:
    for className in classNames:
        print(dataType, " | ", className)
        inputFolder = inputPath + dataType + "/" + className
        data = []
        count = 0
        for root, dirs, files in os.walk(inputFolder):
            for f in files:
                count += 1
                path = os.path.relpath(os.path.join(root, f), ".")
#                 print(count,"\t:: read: ",path)
                data.append(fileHandle(path))
        createBovW(dataType, className)
print("All Done !!!")    
```
