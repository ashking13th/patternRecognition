```python
import numpy as np
import matplotlib.pyplot as plt
import os, argparse, math, random
from datetime import datetime
import matplotlib.patches as mpatches

N = [2,3,4,5]
M = [4,8,16,32]

outputPath = "../output/hmm/accuracy_"

xVals = [4,8,16,32]
yVals = []

myVals = []


# N = 2
yVals.append([36.97, 45.37, 45.37, 47.89])
# N = 3
yVals.append([40.75, 50, 45.79, 50.42])
# N = 4
yVals.append([38.65, 47.05, 46.21, 49.15])
# N = 5
yVals.append([37.81, 46.21, 48.73, 49.15])

myVals.append([36.97, 40.75, 38.65, 37.81])
myVals.append([45.37, 50, 47.05, 46.21])
myVals.append([45.37, 45.79, 46.21, 48.73])
myVals.append([47.89, 50.42, 49.15, 49.15])

colors = ['blue','green', 'orange', 'red']

recs = []
for i in range(len(colors)):
    recs.append(mpatches.Rectangle((0,0),1,1,fc=colors[i]))
```

```python
def plotGraph(n, index, xVal, yVal, xLabel="Accuracy", yLabel="M",filename="", flag=False):
#     if flag:
#         plt.figure()
    plt.plot(xVal, yVal)
    plt.title("Accuracy vs M for N = "+str(n))
    plt.xlabel(xLabel)
    plt.ylabel(yLabel)
#     plt.show()
    plt.savefig(outputPath+filename+str(n)+".jpg")
```

```python
legendText = []
plt.figure()
for i in range(4):
    legendText.append(("N = "+str(N[i])))
    plt.legend(recs, legendText, loc='lower right')
#     print("i: ",i)
    
#     plt.figure()
    plt.plot(M, yVals[i], c = colors[i])
    
    plt.title("Accuracy vs M ")
    plt.ylabel("Accuracy")
    plt.xlabel("M")
#     plt.show()
plt.savefig(outputPath+"M.jpg")
```
