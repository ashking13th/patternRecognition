```python
def forwdProc(obs, n):
    # Assuming I know n
    lObs = len(obs)
    
    alpha = np.zeros((lObs, n))
    for i in range(n):
        alpha[0][i] = PI[i] * B[i][obs[0]]
        
    for ti in range(2,lObs):
        for j in range(n):
            storeVal = 0
            for i in range(n):
                storeVal += alpha[ti-1][i] * A[i][j]
            alpha[ti][j] = storeVal * B[ti][j]
    
    return alpha
```

```python
def backwdProc(obs, n):
    # assuming I know n
    lObs = len(obs)
    beta = np.zeros((lObs, n))
    
    for i in range(n):
        beta[lObs-1][i] = 1
    
    for ti in range(lObs-2, -1, -1):
        for i in range(n):
            storeVal = 0
            for j in range(n):
                storeVal += A[i][j] * B[j][obs[ti+1]] * beta[ti+1][j]
            beta[ti][i] = storeVal
    
    return beta
```
